## Génération du fichier global

```python
filter_types = ["hors-mutation", "mutation+hors-mutation"]

columns = ["Année", "Total", "Chambre", "T1", "T2", "T3", "T4", "T5", "T6 ou plus"]

def get_montrouge_data(input_path):
    df = pd.read_excel(input_path, header=8, usecols='B:L', index_col=[0,1])
    year = input_path.name[:4]
    df["Année"]=year
    return df.loc["Montrouge"]
```

```python
with pd.ExcelWriter(DATA_DIR / "synthese.xlsx") as writer:
    for filter_type in filter_types:
        concat_df = pd.concat([get_montrouge_data(path) for path in (DATA_DIR / filter_type).glob("20*.xlsx")]).dropna(axis='columns', how='all')
        
        concat_df.to_excel(writer, sheet_name=filter_type, columns = columns)
```
