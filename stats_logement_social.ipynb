# Attribution | Demandes satisfaites

```python
SATISFIED_DATA_DIR = Path("data/demandes_satisfaites_logements_sociaux/xlsx/REGION Ile-de-France/")
```

```python
def get_montrouge_satisf_data_legacy(input_path):
    """
    NB: 
    * cette fonction utilise le nom du fichier pour renseigner l'année...
    Cette partie a nécessité un pré-traitement manuel !
    * évidemment, le format a évolué pour les données de 2022 et ultérieur
    """
    print(f"Processing {input_path}")
    df = pd.read_excel(input_path, header=7, usecols='B:M', index_col=[0,1])
    year = input_path.name[:4]
    df["année"]=year
    return df.loc["Montrouge"]
```

```python
def get_montrouge_satisf_data_new(input_path):
    """
    NB: 
    * cette fonction utilise le nom du fichier pour renseigner l'année...
    Cette partie a nécessité un pré-traitement manuel !
    * évidemment, le format a évolué pour les données de 2022 et ultérieur
    """
    print(f"Processing {input_path}")
    df = pd.read_excel(input_path, header=9, usecols='B:M', index_col=[0,1])
    year = input_path.name[:4]
    df["année"]=year
    return df.loc["Montrouge"]
```

```python
satisf_concat_df = pd.concat([
    get_montrouge_satisf_data_legacy(path) for path in (SATISFIED_DATA_DIR / "hors-mutation").glob("*.xlsx") if path.name[:4] in [str(year) for year in range(2015, 2022)]
] + [
    get_montrouge_satisf_data_new(path) for path in (SATISFIED_DATA_DIR / "hors-mutation").glob("*.xlsx") if path.name[:4] in ["2022", "2023"]
]).dropna(axis='columns', how='all')
```
